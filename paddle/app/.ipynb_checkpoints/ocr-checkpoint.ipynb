```python
def leer_data_in_json(path):
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)
    return data
```

```python
def cargar_datos_referencia(json_path):
    """Carga y procesa el JSON de marcas y modelos"""
    json_data = leer_data_in_json(json_path)["rows"]

    marcas_validas = set()
    modelos_por_marca = {}

    for row in json_data:
        marca = row["marca"].upper()
        modelo = row["modelo"].upper()

        marcas_validas.add(marca)
        if marca not in modelos_por_marca:
            modelos_por_marca[marca] = set()
        modelos_por_marca[marca].add(modelo)

    return marcas_validas, modelos_por_marca
```

```python
# Configurar paralelizaci√≥n
num_cores = multiprocessing.cpu_count()
print(f"üñ•Ô∏è  CPUs disponibles: {num_cores}")
pandarallel.initialize(nb_workers=num_cores)

# Habilitar barras de progreso en pandas
tqdm.pandas()

def procesar_excel(input_df, json_path, output_path):
    """Funci√≥n principal que procesa el Excel completo"""
    
    tiempo_inicio = time.time()
    
    # ========== 0. DEBUG: VER COLUMNAS ORIGINALES ==========
    print("=== DEBUG PASO 0: COLUMNAS ORIGINALES ===")
    print(f"Nombres de columnas: {input_df.columns.tolist()}")
    print(f"Cantidad de columnas: {len(input_df.columns)}")
    print("\nPrimeras 10 filas SIN PROCESAR:")
    print(input_df.head(10))
    print("\n")
    
    # ========== 1. CARGAR DATOS DE REFERENCIA ==========
    marcas_validas, modelos_por_marca = cargar_datos_referencia(json_path)
    
    # ========== 1.5 PRE-COMPILAR PATRONES REGEX (OPTIMIZACI√ìN) ==========
    print("=== Pre-compilando patrones regex para b√∫squeda r√°pida ===")
    
    # Pre-compilar patrones de marcas
    patrones_marcas = []
    for marca in sorted(marcas_validas, key=len, reverse=True):
        patron = re.compile(r'\b' + re.escape(marca.lower()) + r'\b')
        patrones_marcas.append((marca, patron))
    
    # Pre-compilar patrones de modelos
    patrones_modelos = {}
    for marca, modelos in modelos_por_marca.items():
        patrones_modelos[marca] = []
        for modelo in modelos:
            patron = re.compile(r'\b' + re.escape(modelo.upper()) + r'\b')
            patrones_modelos[marca].append((modelo, patron, len(modelo)))
    
    print(f"‚úì {len(patrones_marcas)} patrones de marcas compilados")
    print(f"‚úì {sum(len(v) for v in patrones_modelos.values())} patrones de modelos compilados\n")
    
    # ========== 2. COPIAR Y PREPARAR DATAFRAME ==========
    df = input_df.copy()
    
    # AJUSTE: Todo est√° en la columna "Marca", crear estructura correcta
    df['Texto_Original'] = df['Modelo']  # Guardar el texto original
    df['A√±o'] = None
    df['Version'] = None
    df['Llave'] = None
    df['Valor venta'] = None
    df['Valor compra'] = None
    
    # Mantener Valores original si existe
    if 'Valores' not in df.columns:
        df['Valores'] = None
    
    print("=== DEBUG PASO 2: Estructura creada ===")
    print(df[['Texto_Original', 'Marca', 'Modelo']].head(10))
    
    # ========== 4. FUNCI√ìN: ENCONTRAR MARCA (OPTIMIZADA) ==========
    def encontrar_marca_rapido(texto, patrones_marcas):
        """Busca marcas v√°lidas en el texto usando patrones pre-compilados"""
        if pd.isna(texto):
            return None
        
        texto_str = str(texto).strip()
        texto_lower = texto_str.lower()
        
        for marca, patron in patrones_marcas:
            # Coincidencia exacta primero
            if texto_lower == marca.lower():
                return marca
            
            # Palabra completa con patron pre-compilado
            if patron.search(texto_lower):
                return marca
        
        return None
    
    # ========== 5. PROCESAR MARCA (buscar en Texto_Original) ==========
    print("=== Procesando Marcas (paralelo) ===")
    df['Marca_encontrada'] = df['Texto_Original'].parallel_apply(
        lambda x: encontrar_marca_rapido(x, patrones_marcas)
    )
    
    # DEBUG PASO 5: Ver marcas encontradas
    print("\n=== DEBUG PASO 5: MARCAS ===")
    print(df[['Texto_Original', 'Marca_encontrada']].head(30))
    
    # Actualizar Marca
    df['Marca'] = df['Marca_encontrada'].where(df['Marca_encontrada'].notna(), None)
    df['Marca'] = df['Marca'].ffill()
    
    # ========== 6. FUNCI√ìN: ENCONTRAR MODELO Y MARCA (OPTIMIZADA) ==========
    def encontrar_modelo_y_marca_rapido(texto_original, patrones_modelos):
        """Busca modelos v√°lidos usando patrones pre-compilados"""
        if pd.isna(texto_original):
            return None, None
        
        texto_upper = str(texto_original).upper()
        modelos_encontrados = []
        
        for marca, patrones in patrones_modelos.items():
            for modelo, patron, longitud in patrones:
                if patron.search(texto_upper):
                    modelos_encontrados.append((marca, modelo, longitud))
        
        # Retornar el modelo m√°s largo encontrado (m√°s espec√≠fico)
        if modelos_encontrados:
            modelos_encontrados.sort(key=lambda x: x[2], reverse=True)
            return modelos_encontrados[0][0], modelos_encontrados[0][1]
        
        return None, None
    
    # ========== 7. PROCESAR MODELO ==========
    print("\n=== Procesando Modelos (paralelo - esto es lo m√°s lento) ===")
    
    # Crear funci√≥n wrapper para parallel_apply
    def buscar_modelo_wrapper(texto):
        return pd.Series(encontrar_modelo_y_marca_rapido(texto, patrones_modelos))
    
    df[['Marca_por_modelo', 'Modelo_detectado']] = df['Texto_Original'].parallel_apply(
        buscar_modelo_wrapper
    )
    
    # Actualizar Marca si se encontr√≥ por modelo (tiene prioridad)
    df['Marca'] = df['Marca_por_modelo'].where(df['Marca_por_modelo'].notna(), df['Marca'])
    df['Marca'] = df['Marca'].ffill()
    
    # Actualizar Modelo
    df['Modelo'] = df['Modelo_detectado'].where(df['Modelo_detectado'].notna(), df['Modelo'])
    
    # ========== 8. FUNCI√ìN: VALIDAR MODELO ==========
    def es_modelo_valido(texto, modelos_por_marca):
        """Verifica si el texto es un modelo v√°lido"""
        if pd.isna(texto):
            return False
        
        texto_str = str(texto).strip()
        
        # Si es solo n√∫meros, no es v√°lido
        if texto_str.replace('.', '').replace(' ', '').isdigit():
            return False
        
        # Si es muy corto, no es v√°lido
        if len(texto_str) < 3:
            return False
        
        # Si est√° en el diccionario de modelos, s√≠ es v√°lido
        for modelos in modelos_por_marca.values():
            if texto_str.upper() in [m.upper() for m in modelos]:
                return True
        
        return False
    
    # Limpiar Modelo: poner None si no es v√°lido
    df['Modelo'] = df.apply(
        lambda row: row['Modelo'] if es_modelo_valido(row['Modelo'], modelos_por_marca) else None,
        axis=1
    )
    
    # Forward fill del modelo
    df['Modelo'] = df['Modelo'].ffill()
    
    # ========== 9. FUNCI√ìN: ENCONTRAR A√ëO ==========
    def encontrar_a√±o(texto):
        """Busca a√±os de 4 d√≠gitos en el texto"""
        if pd.isna(texto):
            return None
        
        texto_str = str(texto).strip()
        
        # Buscar a√±os de 4 d√≠gitos (2000-2039)
        patron_a√±o = r'\b(20[0-3]\d)\b'
        match = re.search(patron_a√±o, texto_str)
        
        if match:
            return int(match.group(1))
        
        return None
    
    # ========== 10. PROCESAR A√ëO ==========
    print("\n=== Procesando A√±os (paralelo) ===")
    df['A√±o_encontrado'] = df['Texto_Original'].parallel_apply(encontrar_a√±o)
    
    # Actualizar A√±o
    df['A√±o'] = df['A√±o_encontrado'].where(df['A√±o_encontrado'].notna(), df['A√±o'])
    df['A√±o'] = df['A√±o'].ffill()
    
    # ========== 11. PROCESAR VERSION Y VALOR VENTA ==========
    print("\n=== Procesando Version y Valor Venta (paralelo) ===")
    def extraer_version_y_valor(texto):
        """Divide el texto en Version y Valor venta usando Man/Aut como separador"""
        if pd.isna(texto):
            return None, None
        
        texto_str = str(texto).strip()
        
        # Buscar "Man" o "Aut" con un espacio o n√∫mero despu√©s
        patron = r'\b(Man|Aut)\b\s+(\d+)'
        match = re.search(patron, texto_str)
        
        if match:
            # Posici√≥n donde termina "Man" o "Aut"
            pos_fin_transmision = match.end(1)
            # Valor est√° despu√©s del espacio
            valor = match.group(2)
            
            # Version es todo hasta el final de Man/Aut (inclusive)
            version = texto_str[:pos_fin_transmision].strip()
            
            return version, valor
        
        # Si no encuentra Man/Aut, todo va a Version
        return texto_str, None
    
    # Aplicar la extracci√≥n
    df[['Version', 'Valor venta']] = df['Texto_Original'].parallel_apply(
        lambda x: pd.Series(extraer_version_y_valor(x))
    )
    
    # ========== 12. LIMPIAR VERSION (quitar Marca, Modelo y A√±o) ==========
    print("\n=== Limpiando Version (paralelo) ===")
    def limpiar_version(row):
        """Remueve Marca, Modelo y A√±o de la columna Version"""
        version = row['Version']
        
        if pd.isna(version):
            return None
        
        version_limpia = str(version)
        
        # Remover Marca
        if pd.notna(row['Marca']):
            marca = str(row['Marca'])
            patron_marca = r'\b' + re.escape(marca) + r'\b'
            version_limpia = re.sub(patron_marca, '', version_limpia, flags=re.IGNORECASE)
        
        # Remover Modelo
        if pd.notna(row['Modelo']):
            modelo = str(row['Modelo'])
            patron_modelo = r'\b' + re.escape(modelo) + r'\b'
            version_limpia = re.sub(patron_modelo, '', version_limpia, flags=re.IGNORECASE)
        
        # Remover A√±o
        if pd.notna(row['A√±o']):
            a√±o = str(int(row['A√±o']))
            patron_a√±o = r'\b' + a√±o + r'\b'
            version_limpia = re.sub(patron_a√±o, '', version_limpia)
        
        # Limpiar espacios m√∫ltiples y espacios al inicio/final
        version_limpia = re.sub(r'\s+', ' ', version_limpia).strip()
        
        return version_limpia if version_limpia else None
    
    df['Version'] = df.parallel_apply(limpiar_version, axis=1)
    
    print("\n=== DEBUG PASO 12: VERSION LIMPIA ===")
    print(df[['Marca', 'Modelo', 'A√±o', 'Version', 'Valor venta']].head(20))
    
    # ========== 13. LIMPIAR COLUMNAS TEMPORALES ==========
    df = df.drop(['Marca_encontrada', 'Marca_por_modelo', 'Modelo_detectado', 'A√±o_encontrado'], axis=1)
    
    # ========== 14. REORDENAR COLUMNAS ==========
    columnas_orden = ["Marca", "Modelo", "A√±o", "Version", "Llave", "Valor venta", "Valor compra", "Texto_Original"]
    df = df[columnas_orden]
    
    # ========== 15. RESUMEN FINAL ==========
    tiempo_total = time.time() - tiempo_inicio
    
    print("\n" + "="*60)
    print("‚úì PROCESAMIENTO COMPLETADO")
    print("="*60)
    print(f"üñ•Ô∏è  CPUs utilizados: {num_cores}")
    print(f"‚è±Ô∏è  Tiempo total: {tiempo_total:.2f} segundos ({tiempo_total/60:.2f} minutos)")
    print(f"üìä Total de registros procesados: {len(df):,}")
    print(f"üè≠ Marcas √∫nicas detectadas: {df['Marca'].nunique()}")
    print(f"üöó Modelos √∫nicos detectados: {df['Modelo'].nunique()}")
    print(f"üìÖ A√±os √∫nicos detectados: {df['A√±o'].nunique()}")
    print(f"üí∞ Valores detectados: {df['Valor venta'].notna().sum():,}")
    print(f"‚ö° Velocidad: {len(df)/tiempo_total:.0f} registros/segundo")
    print(f"üöÄ Aceleraci√≥n: ~{num_cores}x m√°s r√°pido que un solo n√∫cleo")
    print("="*60 + "\n")
    
    # ========== 16. GUARDAR ARCHIVO EXCEL ==========
    if output_path:
        print(f"üíæ Guardando archivo en: {output_path}")
        try:
            # Guardar con formato y estilos
            df.to_excel(
                output_path, 
                index=False,           # Sin √≠ndice
                sheet_name='Datos',    # Nombre de la hoja
                engine='openpyxl'      # Motor recomendado
            )
            print(f"‚úÖ Archivo guardado exitosamente!\n")
        except Exception as e:
            print(f"‚ùå Error al guardar archivo: {e}\n")
            print("üí° Tip: Aseg√∫rate de tener instalado openpyxl: pip install openpyxl\n")
    
    return df
```

```python
# EJECUCI√ìN PRINCIPAL

resultado = procesar_excel(
    input_df=final_df,
    json_path="data.json",
    output_path=excel_path
)
```
